```python
q1 = df.quantile(0.25)
q3 = df.quantile(0.75)
```

```python
IQR = q3 - q1
```

```python
IQR
```

```python
### mengambil / mengeluarkan / menandai poin data yang merupakan outlier

(df < (q1 - 1.5 * IQR)) | (df > (q3 + 1.5 * IQR))
```

```python
df_no_outlier = df[~((df < (q1 - 1.5 * IQR)) | (df > (q3 + 1.5 * IQR))).any(axis=1)]
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
df.head()
```

```python
df.columns
```

```python
xfeatures = df[["age","sex","steroid","antivirals","fatigue","malaise",
               "anorexia","liver_big","liver_firm","spleen_palpable","spiders",
               "ascites","varices","bilirubin","alk_phosphate","sgot","albumin",
               "protime","histology"]]
ylabels = df["class"]
```

```python
# menggunakan Skbest
skb = SelectKBest(score_func=chi2, k=10)
best_feature_fit = skb.fit(xfeatures,ylabels)
```

```python
# skor terbaik menunjukkan fit terbaik
print("Skor", best_feature_fit.scores_)
```

```python
# Transform (pilihan)
bf_02 = best_feature_fit.transform(xfeatures)
```

```python
bf_02
```

```python
# melakukan mapping ke namaa fitur
feature_scores = pd.DataFrame(best_feature_fit.scores_, columns=["Feature_Scores"])
```

```python
feature_scores
```

```python
feature_column_names = pd.DataFrame(xfeatures.columns, columns=["Feature_name"])
best_feat_df = pd.concat([feature_scores,feature_column_names],axis=1)
```

```python
best_feat_df
```

```python
# mendapatkan yang terbaik 
# semakin besar angkanya semaking penting fiturnya
best_feat_df.nlargest(10, "Feature_Scores")
```

```python
best_feat_df.nlargest(12, "Feature_Scores")
```

```python
# package ML
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

# Metriks
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
# fitur untuk data awal
df.columns
```

```python
xfeatures
```

```python
ylabels = df['class']
```

```python
# fitur terpilih
df.columns
```

```python
xfeatures_best = df[["age", "sex", "steroid", "antivirals", "fatigue", "spiders",
                    "ascites", "varices", "bilirubin", "alk_phosphate", "sgot", "albumin",
                    "protime", "histology"]]
```

```python
# train / test dataset
x_train, x_test, y_train, y_test = train_test_split(xfeatures,ylabels,test_size=0.3,random_state=7)
```

```python
# train / test dataset untuk fitur terbaik
x_train_b,x_test_b,y_train_b,y_test_b = train_test_split(xfeatures_best,ylabels,test_size=0.30,random_state=7)
```

```python
# membangun model
logreg = LogisticRegression()
logreg.fit(x_train,y_train)
```

```python
# Akurasi Model
# Metode 1
logreg.score(x_test,y_test)
```

```python
# Metode 2
logreg.predict(x_test)
```

```python
print(list(y_test))
```

```python
accuracy_score(y_test,logreg.predict(x_test))
```

```python
# Membangun model
model_logit = LogisticRegression()
model_logit.fit(x_train_b, y_train_b)
```

```python
# Akurasi Model
# Metode 1
model_logit.score(x_test_b,y_test_b)
```

```python
### Prediksi tunggal
x_test.iloc[1]
```
